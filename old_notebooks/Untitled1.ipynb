```python
import networkx as nx

import stormvogel.examples as examples
from stormvogel.graph import ModelGraph
from stormvogel.visualization import JSVisualization, MplVisualization
from stormvogel.layout import DEFAULT
from stormvogel.show import show


mdp = examples.create_lion_mdp()
G = ModelGraph.from_model(mdp)
pos = nx.bfs_layout(G, start=mdp.get_initial_state().id)

layout = DEFAULT().set_nx_pos(pos)
vis = show(mdp, layout=layout, show_editor=True)
```

```python
layout = DEFAULT().set_nx_pos(pos)
layout.set_value(["groups", "states", "color", "background"], "black")
MplVisualization(mdp, layout=layout).show()
```
