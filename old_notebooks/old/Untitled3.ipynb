```python
from IPython.display import display, HTML, Javascript

# JavaScript to detect when the user scrolls to the cell
scroll_alert_script = """
  alert('starting now');
  // Wait for the document to be fully loaded
  window.addEventListener('load', function() {
      const targetCell = document.querySelector('.cell'); // Select the cell element

      // Create an intersection observer to detect when the cell enters the viewport
      const observer = new IntersectionObserver(function(entries, observer) {
          entries.forEach(entry => {
              if (entry.isIntersecting) {
                  alert('You have scrolled to this cell!'); // Trigger alert when the cell is visible
                  observer.disconnect(); // Disconnect the observer after it triggers the alert
              }
          });
      }, { threshold: 0.5 }); // Trigger when 50% of the cell is visible

      // Start observing the target cell
      observer.observe(targetCell);
  });
"""

# Inject JavaScript code into the Jupyter notebook
display(Javascript(scroll_alert_script))
```

```python
from IPython.display import display, Javascript, HTML

# JavaScript to trigger alert when the user scrolls to the cell
scroll_alert_script = """
alert('new');
  // Wait for the document to be fully loaded
  window.addEventListener('load', function() {
      // Get the current cell (this one that is running the code) using the unique data-cell-id
      var cell = document.querySelector('.cell');
      if (!cell) {
          alert('Cell not found');
          return;
      }

      // Create an intersection observer to detect when the cell enters the viewport
      var observer = new IntersectionObserver(function(entries, observer) {
          entries.forEach(entry => {
              if (entry.isIntersecting) {
                  alert('You have scrolled to this cell!'); // Show alert when the cell is in view
                  observer.disconnect(); // Disconnect after it triggers to prevent multiple alerts
              }
          });
      }, { threshold: 0.5 }); // Trigger when 50% of the cell is visible

      // Start observing the current cell
      observer.observe(cell);
  });
"""

# Inject JavaScript code into the Jupyter notebook
display(HTML("<p>hi</p><script>" + scroll_alert_script + "</script>"))
```
